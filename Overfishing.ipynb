```python
# set size figure
sns.set(rc={'figure.figsize': (12,10)})
```

```python
sns.lineplot(x = "Year", y = "Catch_Tons", data = us, hue='Species')
plt.title ('Unites States Catch of Fish by Species', size=15)
```

```python
# the overfishing of the atlantic cod catch overpower the other species metrics 

# create new us df without cod 

us_no_cod = df[(df.Country == 'United States of America') & (df.Species != 'Atlantic cod')]
```

```python
# new US line graph without cod
ax = sns.lineplot(x='Year', y='Catch_Tons', data=us_no_cod, hue='Species', ci=False)
plt.title('Unites States Catch of Fish by Species (Cod removed)', size=15)
plt.show()
```

#### Obervations 
* bluefin tuna accounted major of the catch in the 60s and 70s followed by dramatic drops 
* swordfish catch reached record highs in 1990 followed by dramatic drop
* halibut had few peaks followed by the drops that now seem to follow climbs
* only halibut appears to have been steach catch from 50s to 2020 
* it appears that overfishing of species is followed by drop this is know as tragedy of the commons - where is a shared environment - parties act in their own interest to the detriment of the common good

```python
# create lineplot of timeline for each individual data species subframe
sns.lineplot(x = "Year", y = "Catch_Tons", data = bluefin, hue='Country', ci=False)
plt.title ('1950-2022 Bluefin Tuna Catch', size=15)
plt.show()
```

#### Bluefin observations 
* Norway dramatic overfishing 50s and 60s followed by dramatic drop. By the 1990s Norways catch of the blue fin was near zero - this could be bec it has fallen out of favor but more likely due to regulation
* Ireland does not favor the blue fin - fun fact
* US has more moderate overfishing spikes followed by drops 
* Canada had a more moderate version of US bluefincatch 


```python
sns.lineplot(x = "Year", y = "Catch_Tons", data = cod, hue='Country', ci=False)
plt.title ('1950-2022 Cod Catch', size=15)
plt.show()
```

#### Observations
* Who knew there was a cod craze? Canada Iceland and Norway all competed for this species over timeline
* it seems even though the patern was repeating in the drops the behavior of overfishing continues
* only after Canadas 500k ton peak droped 500X did we see a reduction in canada cod fishing
* it appears that Norway and Iceland are still overfishing like its 1980!!
* Portugual, US, UK, Ireland play less of a role on the cod craze

```python
sns.lineplot(x = "Year", y = "Catch_Tons", data = halibut, hue='Country', ci=False)
plt.title ('1950-2022 Halibut Catch', size=15)
plt.show()
```

#### Observations

```python
sns.lineplot(x = "Year", y = "Catch_Tons", data = swordfish, hue='Country', ci=False)
plt.title ('1950-2022 Swordfish Catch', size=15)
plt.show()
```

```python
sns.lineplot(x = "Year", y = "Catch_Tons", data = salmon, hue='Country', ci=False)
plt.title ('1950-2022 Salmon Catch', size=15)
plt.show()
```

#### Obervations
